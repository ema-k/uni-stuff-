```python
from scipy.integrate import odeint
from scipy.optimize import brentq
from scipy.integrate import quad
from numpy import *
from pylab import *
from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt

firebrick=(178./255.,34./255.,34./255.)
indianred=(176./255.,23./255.,31./255.)
steelblue=(70./255.,130./255.,180./255.)
slategray1=(198./255.,226./255.,255./255.)
slategray4=(108./255.,123./255.,139./255.)
lavender=(230./255.,230./255.,230./255.)
cobalt=(61./255.,89./255.,171./255.)
midnightblue=(25./255.,25./255.,112./255.)
forestgreen=(34./255.,139./255.,34./255.)

alpha=2.0
m=1.0
E=1.5
U0=1.0
x0=math.pi/(2*alpha)

#the goal is to compare exact and analytical solution (approximation) of a potential

def potencijal(x):
    return U0/((np.sin(alpha*x))**2)
def sila(x):
        return (2*alpha*U0*np.cos(alpha*x))/(np.sin(alpha*x))**3
def funct(y,t):
    x = y[0]
    return (y[1],sila(x)/m)

#initial position and velocity
x0 = np.pi/(2*alpha)
dt_x0 = np.sqrt(2/m)*np.sqrt(E-U0)

tmax = 50.0
dt = tmax/100
t = np.arange(0.0,tmax+dt,dt)

y0 = [x0,dt_x0]
y=odeint(funct,y0,t) 


def xegz(t):
    xe = (1/alpha)*np.arccos(np.sqrt(1-U0/E)*np.sin(-alpha*np.sqrt(2*E/m)*t))
    return xe

def vxegz(t):
    vxe = (np.sqrt((2*E-U0)/m)*np.cos(alpha*np.sqrt(2*E/m)*t))/np.sqrt(1-(1-U0/E)*np.sin(alpha*np.sqrt(2*E/m)*t))
    return vxe

fig1=plt.figure(1)
plt.plot(t,y[:,0],color=firebrick,ls='--',lw=1.5,label='numerical')
plt.plot(t,xegz(t),color=steelblue,ls='-',lw=1,label='analytical')
plt.legend()

fig1=plt.figure(2)
plt.plot(t,y[:,1],color=firebrick,ls='--',lw=1.5,label='numerical')
plt.plot(t,vxegz(t),color=steelblue,ls='-',lw=1,label='analytical')
plt.legend()
    
plt.show()
```
